```python
from collections import defaultdict

sharpness_levels = [64, 79, 90, 87, 73, 63, 76, 89]

def process_joint_data(file_paths, sharpness_levels):
    """Process acceleration data from multiple Excel files with different sharpness levels."""
    # Dictionary to store processed data by sheet type
    sheet_data = defaultdict(list)
    
    # Process all files and extract data by sheet type
    for file_path, sharpness in zip(file_paths, sharpness_levels):
        try:
            xls = pd.ExcelFile(file_path)
            
            for sheet_name in xls.sheet_names:
                if sheet_name not in JOINT_SHEET_XZY:
                    continue
                
                try:
                    # Read the sheet
                    df = pd.read_excel(xls, sheet_name=sheet_name)
                    print(f"{df.shape}")
                    
                    if df.empty:
                        continue
                    
                    # Add sharpness column
                    df['sharpness'] = sharpness
                    
                    # Add sheet name as a column to differentiate data source
                    df['sheet_type'] = sheet_name
                    
                    # Add to our collection
                    sheet_data[sheet_name].append(df)
                
                except Exception as e:
                    print(f"Error processing sheet {file_path} {sheet_name}: {e}")
        
        except Exception as e:
            print(f"Error processing file {file_path}: {e}")
    
    # Combine data from all files
    combined_dfs = []
    
    for sheet_name, dfs in sheet_data.items():
        if dfs:
            # Concatenate all data for this sheet type
            sheet_combined_df = pd.concat(dfs, ignore_index=True)
            combined_dfs.append(sheet_combined_df)
    
    # Return combined data frame
    if combined_dfs:
        final_df = pd.concat(combined_dfs, ignore_index=True)
        return final_df
    
    return None

def main():
    # Process the data
    merged_df = process_joint_data(
        file_paths=BONING_PATHS,
        sharpness_levels=sharpness_levels
    )
    
    if merged_df is None or merged_df.empty:
        print("No results generated!")
        return
    
    # Save the final merged dataframe
    output_file = "boning_joint_data_XZY.csv"
    merged_df.to_csv(output_file, index=False)
    print(f"Saved data to {output_file} ({merged_df.shape[0]} rows, {merged_df.shape[1]} columns)")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import pandas as pd
from collections import Counter

from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split

from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import TomekLinks
from imblearn.pipeline import Pipeline as ImbPipeline

def balance_sharpness(df,
                      sharp_col='sharpness',
                      test_size=0.3,
                      val_size=0.5,
                      target_fraction=0.5,
                      random_state=42):
    # 1) train / (val+test) split, stratified by sharpness
    train_df, temp_df = train_test_split(
        df, test_size=test_size,
        stratify=df[sharp_col],
        random_state=random_state
    )
    # 2) split temp into val / test
    val_df, test_df = train_test_split(
        temp_df, test_size=val_size,
        stratify=temp_df[sharp_col],
        random_state=random_state
    )

    # 3) prepare X and y for training
    #    drop both 'Label' and the target column
    X = train_df.drop(columns=['Label', sharp_col], errors='ignore').reset_index(drop=True)
    y = train_df[sharp_col].reset_index(drop=True)

    # 4) scale numeric features
    numeric_feats = X.select_dtypes(include=[np.number]).columns.tolist()
    scaler = RobustScaler()
    X[numeric_feats] = scaler.fit_transform(X[numeric_feats])

    # 5) build SMOTE sampling_strategy to bring everyone to
    #    `target_fraction` of the largest class
    counts = Counter(y)
    majority = max(counts.values())
    target_n = int(target_fraction * majority)
    smote_strat = {
        cls: target_n
        for cls, cnt in counts.items()
        if cnt < target_n
    }

    # 6) pipeline: SMOTE then TomekLinks
    sm = SMOTE(
        sampling_strategy=smote_strat,
        random_state=random_state,
        k_neighbors=5
    )
    tl = TomekLinks(sampling_strategy='all')
    pipeline = ImbPipeline([('smote', sm), ('tomek', tl)])

    # 7) fit_resample
    X_res, y_res = pipeline.fit_resample(X, y)

    # 8) rebuild balanced DataFrame
    balanced_train = pd.concat([
        pd.DataFrame(X_res, columns=X.columns),
        pd.Series(y_res, name=sharp_col)
    ], axis=1)

    # 9) sanity‐check counts
    print("New sharpness distribution:")
    print(balanced_train[sharp_col].value_counts().sort_index())

    return balanced_train, val_df.reset_index(drop=True), test_df.reset_index(drop=True)
```

# **Data Cleaning and minimize extreme values by replacing it with its inter-quartile ranges**

```python
df = balanced_train

num_cols = df.select_dtypes(include='number').columns

Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

mask = ~((df[num_cols] < (Q1 - 1.5 * IQR)) |
         (df[num_cols] > (Q3 + 1.5 * IQR))).any(axis=1)

df_iqr_clean = df[mask].reset_index(drop=True)

print(f"Kept {len(df_iqr_clean)}/{len(df)} rows after IQR filtering")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# ————————————————————————————————
# 1) (Re-)load your DataFrame
# assume df is already in your workspace; otherwise:
# df = pd.read_csv('/path/to/your_data.csv')
df = df_iqr_clean
# ————————————————————————————————

levels = sorted(df['sharpness'].unique())

# ————————————————————————————————
# 2) Compute per-part metrics
# ————————————————————————————————

# Elbow
elbow = {
    'Right Elbow Flexion Reps': [],
    'Left Elbow Flexion Reps': [],
    'Right Elbow Extreme Time': [],
    'Left Elbow Extreme Time': [],
    'Right Elbow Variability': [],
    'Left Elbow Variability': []
}

# Shoulder
shoulder = {
    'Right Shoulder Elevation Time': [],
    'Left Shoulder Elevation Time': [],
    'Right Shoulder Load Time': [],
    'Left Shoulder Load Time': [],
    'Right Shoulder Smoothness': [],
    'Left Shoulder Smoothness': []
}

# Neck & Back
neck_back = {
    'Trunk Flexion Duration': [],
    'Head Tilt Variability': [],
    'Cumulative Trunk Load': []
}

# Whole-body proxies
whole_body = {
    'Movement Asymmetry': [],
    'Energy Expenditure Proxy': []
}

for lvl in levels:
    d = df[df['sharpness']==lvl]
    n = len(d)
    
    #### Elbow ####
    for side in ('Right','Left'):
        sig = d[f'{side} Elbow Flexion/Extension'].values
        # 1) count peaks
        peaks,_ = find_peaks(sig)
        elbow[f'{side} Elbow Flexion Reps'].append(len(peaks))
        # 2) extreme-angle time (> mean+2σ)
        thr = sig.mean() + 2*sig.std()
        elbow[f'{side} Elbow Extreme Time'].append((np.abs(sig)>thr).sum()/n)
        # 3) motion variability (mean STD of 3 DOFs)
        mv = d[[f'{side} Elbow Ulnar Deviation/Radial Deviation',
                f'{side} Elbow Pronation/Supination',
                f'{side} Elbow Flexion/Extension']].std().mean()
        elbow[f'{side} Elbow Variability'].append(mv)
    
    #### Shoulder ####
    for side in ('Right','Left'):
        sig = d[f'{side} Shoulder Flexion/Extension']
        thr = sig.mean() + 2*sig.std()
        t_elev = (sig>thr).sum()/n
        shoulder[f'{side} Shoulder Elevation Time'].append(t_elev)
        shoulder[f'{side} Shoulder Load Time'].append(t_elev)  # same threshold logic
        jerk = np.diff(sig.values, n=2)
        shoulder[f'{side} Shoulder Smoothness'].append(np.std(jerk))
    
    #### Neck & Back ####
    tf = d['L5S1 Flexion/Extension']
    thr_tf = tf.mean() + 2*tf.std()
    neck_back['Trunk Flexion Duration'].append((tf>thr_tf).sum()/n)
    hv = d[['C1 Head Lateral Bending','C1 Head Axial Rotation','C1 Head Flexion/Extension']].std().mean()
    neck_back['Head Tilt Variability'].append(hv)
    neck_back['Cumulative Trunk Load'].append(np.sum(np.abs(tf)))
    
    #### Whole Body ####
    # asymmetry: abs(mean right elbow – mean left)
    re = d['Right Elbow Flexion/Extension'].mean()
    le = d['Left Elbow Flexion/Extension'].mean()
    whole_body['Movement Asymmetry'].append(abs(re-le))
    # energy proxy: sum of variances of frame-to-frame diffs
    vel = d.diff().dropna()
    whole_body['Energy Expenditure Proxy'].append(vel.var().sum())

# ————————————————————————————————
# 3) Build DataFrames
# ————————————————————————————————
df_elbow      = pd.DataFrame(elbow,      index=levels)
df_shoulder   = pd.DataFrame(shoulder,   index=levels)
df_neck_back  = pd.DataFrame(neck_back,  index=levels)
df_whole_body = pd.DataFrame(whole_body, index=levels)

# ————————————————————————————————
# 4) Plot each body part separately
# ————————————————————————————————

def plot_part(df_part, title):
    fig, axs = plt.subplots(len(df_part.columns), 1, figsize=(8, 2.5*len(df_part.columns)), sharex=True)
    for ax, col in zip(axs, df_part.columns):
        ax.plot(df_part.index, df_part[col], marker='o')
        ax.set_ylabel(col)
    axs[-1].set_xlabel('Sharpness Level')
    fig.suptitle(title, y=1.02)
    plt.tight_layout()
    plt.show()

plot_part(df_elbow,     'Elbow Metrics')
plot_part(df_shoulder,  'Shoulder Metrics')
plot_part(df_neck_back, 'Neck & Back Metrics')
plot_part(df_whole_body,'Whole-Body Proxies')
df.info()
```

```python
import pandas as pd
from scipy.stats import f_oneway
from collections import defaultdict

feature_groups = {
    'elbow': ['Elbow'],
    'shoulder': ['Shoulder', 'T4 Shoulder'],
    'neck_back': ['T1C7', 'T9T8', 'L1T12', 'L4L3', 'L5S1', 'C1 Head'],
    'whole_body': ['Hip', 'Knee', 'Ankle', 'Ball Foot', 'Wrist']
}

grouped_features = defaultdict(list)

def stat_comparison(df):    
    for feature in df.columns:
        if feature == 'sharpness':
            continue
        matched = False
        for group_name, keywords in feature_groups.items():
            if any(kw in feature for kw in keywords):
                grouped_features[group_name].append(feature)
                matched = True
                break
    
        if not matched:
            grouped_features['unclassified'].append(feature)
    
    
    sharpness_levels = sorted(df['sharpness'].unique())
    results_by_group = {}
    
    for group, features in grouped_features.items():
        stats_summary = []
    
        for feature in features:
            row = {'Feature': feature}
    
            for level in sharpness_levels:
                values = df[df['sharpness'] == level][feature]
                row[f'Mean_{level}'] = values.mean()
                row[f'Std_{level}'] = values.std()
    
            groups = [df[df['sharpness'] == level][feature].dropna() for level in sharpness_levels]
            if all(len(g) > 1 for g in groups):
                _, p_value = f_oneway(*groups)
                row['ANOVA_p'] = p_value
            else:
                row['ANOVA_p'] = None
    
            stats_summary.append(row)
    
        results_by_group[group] = pd.DataFrame(stats_summary)
    
    
    all_results = []
    
    for group, summary_df in results_by_group.items():
        summary_df['Group'] = group
        summary_df['Significant (p <0.05)'] = summary_df['ANOVA_p'] < 0.05
        all_results.append(summary_df)
    
    final_df = pd.concat(all_results, ignore_index=True)
    
    final_df.sort_values(by=['Group', 'ANOVA_p'])

    return final_df


boning_ZXY_stat = stat_comparison(boning_df_ZXY)
```

```python
import matplotlib.pyplot as plt

def plot_feature_group_means(final_df):
    # sharpness levels are preserved in the column header
    sharpness_levels = sorted({int(col.split('_')[1]) for col in final_df.columns if col.startswith('Mean_')})
    
    feature_groups = final_df['Group'].unique()

    for group in feature_groups:
        group_df = final_df[final_df['Group'] == group]

        means_per_sharpness = {}
        for level in sharpness_levels:
            col = f"Mean_{level}"
            if col in group_df.columns:
                means_per_sharpness[level] = group_df[col].mean()

        plt.figure(figsize=(8,5))
        plt.plot(
            list(means_per_sharpness.keys()),
            list(means_per_sharpness.values()),
            marker='o'
        )

        plt.title(f"Average {group.upper()} Movement vs Knife Sharpness Level")
        plt.xlabel("Knife Sharpness Level")
        plt.ylabel("Average Sensor Value")
        plt.grid(True)
        plt.tight_layout()
        plt.show()
```

2. Sharpness 89–90 shows unexpected drops/spikes:

3. These plots show that knife sharpness has a non-linear effect on body mechanics — which is valuable insight for injury prevention and ergonomic optimization.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def plot_feature_group_violins_with_annotations(raw_df: pd.DataFrame, final_df: pd.DataFrame):
    # 1) grab sharpness levels from your Mean_ columns
    mean_cols = [c for c in final_df.columns if c.startswith("Std_")]
    sharpness_levels = sorted(int(c.split("_")[1]) for c in mean_cols)

    # 2) build a dict of group → feature list
    feature_groups = {
        grp: final_df.loc[final_df["Group"] == grp, "Feature"].tolist()
        for grp in final_df["Group"].unique()
    }

    # 3) loop each group
    for group, feats in feature_groups.items():
        # keep only those features actually in raw_df
        valid_feats = [f for f in feats if f in raw_df.columns]
        if not valid_feats:
            continue

        # melt to long form
        melted = raw_df[["sharpness", *valid_feats]].melt(
            id_vars="sharpness",
            value_vars=valid_feats,
            var_name="Feature",
            value_name="Value",
        )
        # ensure categorical order
        melted["sharpness"] = pd.Categorical(
            melted["sharpness"].astype(int),
            categories=sharpness_levels,
            ordered=True,
        )

        # compute medians per level
        medians = (
            melted.groupby("sharpness")["Value"]
            .median()
            .reindex(sharpness_levels)
        )
        peak_lvl, dip_lvl = medians.idxmax(), medians.idxmin()
        peak_val, dip_val = medians.max(), medians.min()

        # draw violin
        plt.figure(figsize=(8, 5))
        ax = sns.violinplot(
            data=melted,
            x="sharpness",
            y="Value",
            order=sharpness_levels,
            inner="quartile",
            scale="width",
        )

        # annotate: map sharpness → its index
        peak_idx = sharpness_levels.index(peak_lvl)
        dip_idx = sharpness_levels.index(dip_lvl)

        ax.scatter(
            peak_idx,
            peak_val,
            color="red",
            s=100,
            zorder=10,
            label=f"Peak median ({peak_lvl})",
        )
        ax.text(
            peak_idx,
            peak_val + 0.05,
            str(peak_lvl),
            color="red",
            ha="center",
            va="bottom",
            fontweight="bold",
        )

        ax.scatter(
            dip_idx,
            dip_val,
            color="blue",
            s=100,
            zorder=10,
            label=f"Dip median ({dip_lvl})",
        )
        ax.text(
            dip_idx,
            dip_val - 0.05,
            str(dip_lvl),
            color="blue",
            ha="center",
            va="top",
            fontweight="bold",
        )

        # tidy up axes
        ax.set_title(f"{group.upper()} Features Distribution by Sharpness")
        ax.set_xlabel("Knife Sharpness Level")
        ax.set_ylabel("Sensor Value")
        ax.set_xticks(range(len(sharpness_levels)))
        ax.set_xticklabels(sharpness_levels, rotation=45)
        ax.legend(loc="upper right")
        plt.tight_layout()
        plt.show()
```
